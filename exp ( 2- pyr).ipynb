```python
import cv2
import matplotlib.pyplot as plt

# ---------- Read Image ----------
img = cv2.imread("/anime-cityscape-sunset_23-2151975108.jpg")

if img is None:
    raise FileNotFoundError("Image not found. Check file name carefully.")

# Convert BGR to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# ---------- Generate Pyramid Levels ----------
level1 = cv2.pyrDown(img)
level2 = cv2.pyrDown(level1)
level3 = cv2.pyrDown(level2)

images = [img, level1, level2, level3]
titles = ["Level 0 (Original)", "Level 1", "Level 2", "Level 3"]

# ---------- Show Pyramid ----------
plt.figure(figsize=(10,6))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()

# ---------- Print Image Sizes ----------
for i, im in enumerate(images):
    print(f"Level {i} image size: {im.shape}")

# ---------- Edge Detection ----------
for i, im in enumerate(images):
    gray = cv2.cvtColor(im, cv2.COLOR_RGB2GRAY)
    edges = cv2.Canny(gray, 100, 200)

    plt.figure(figsize=(4,4))
    plt.imshow(edges, cmap="gray")
    plt.title(f"Edges at Pyramid Level {i}")
    plt.axis("off")
    plt.show()
```
